# Data Cleaning and Filtering Operations.

1. Remove rows with missing values in the 'Alma_Mater' column using dropna() and applies the changes directly to the DataFrame 'df' with inplace=True.
2. Filter rows where 'Death_Mission' column is null and 'Death_Date' column is not null using the .index method. Then, remove those rows from the DataFrame 'df' by using the .drop() function with the inplace=True parameter.
3. Use pd.to_datetime() with the dayfirst=True parameter to convert the 'Birth_Date' and 'Death_Date' columns to datetime format with day-first parsing separately.

```python
df.dropna(subset=['Alma_Mater'], inplace=True)
rows_to_drop = df[df['Death_Mission'].isnull() & df['Death_Date'].notnull()].index
```

```python
rows_to_drop
```

```python
df.drop(rows_to_drop, inplace=True)
df['Birth_Date'] = pd.to_datetime(df['Birth_Date'], dayfirst=True)
df['Death_Date'] = pd.to_datetime(df['Death_Date'], dayfirst=True)
```

# Now, we will do analysis using SQL:

```python
%sql sqlite:///my_database.sqlite
%sql SELECT * FROM astronauts;
```

This code block retrieves all of the data from the astronauts table. This could be useful for getting a general overview of all of the astronauts in the database, or for identifying specific astronauts who meet certain criteria. 

```python
%sql SELECT Status, COUNT(*) AS Number FROM astronauts GROUP BY Status;
```

This code block groups the astronauts by their status and counts the number of astronauts in each group. This could be useful for understanding how many astronauts are active, retired, or deceased.

```python
%sql SELECT Military_Branch, COUNT(*) AS Number FROM astronauts GROUP BY Military_Branch;
```

This code block groups the astronauts by their military branch and counts the number of astronauts in each group. This could be useful for understanding which military branches are most represented among astronauts.

```python
%sql SELECT Military_Rank, COUNT(*) AS Number FROM astronauts GROUP BY Military_Rank ORDER BY Number DESC LIMIT 5;
```

This code block groups the astronauts by their military rank and counts the number of astronauts in each group, ordered by the number of astronauts in each group in descending order. This could be useful for identifying the most common military ranks among astronauts.

```python
%sql SELECT Gender, COUNT(*) AS Number FROM astronauts GROUP BY Gender;
```

This code block groups the astronauts by their gender and counts the number of astronauts in each group. This could be useful for understanding the gender distribution among astronauts.

```python
%%sql
SELECT ROUND(AVG(life_Expectancy)) AS Average_Life_Expectancy
FROM (
    SELECT
        CASE
            WHEN Status = 'Deceased' THEN strftime('%Y', Death_Date) - strftime('%Y', Birth_Date)
            ELSE 2023 - strftime('%Y', Birth_Date)
        END AS life_Expectancy
    FROM astronauts
) AS LifeExpectancySubquery;
```

This code block calculates the average life expectancy of all astronauts. This could be useful for understanding the overall health and well-being of astronauts.

```python
%%sql
SELECT ROUND(AVG(Female_life_Expectancy)) AS Female_Average_Life_Expectancy
FROM (
    SELECT
        CASE
            WHEN Status = 'Deceased' AND Gender = 'Female' THEN strftime('%Y', Death_Date) - strftime('%Y', Birth_Date)
            WHEN Status != 'Deceased' AND Gender = 'Female' THEN 2023 - strftime('%Y', Birth_Date)
            ELSE NULL
        END AS Female_life_Expectancy
    FROM astronauts
) AS FemaleLifeExpectancySubquery;
```

This code block calculates the average life expectancy of female astronauts. This could be useful for understanding the gender differences in life expectancy among astronauts.

```python
%%sql
SELECT ROUND(AVG(Male_life_Expectancy)) AS Male_Average_Life_Expectancy
FROM (
    SELECT
        CASE
            WHEN Status = 'Deceased' AND Gender = 'Male' THEN strftime('%Y', Death_Date) - strftime('%Y', Birth_Date)
            WHEN Status != 'Deceased' AND Gender = 'Male' THEN 2023 - strftime('%Y', Birth_Date)
            ELSE NULL
        END AS Male_life_Expectancy
    FROM astronauts
) AS MaleLifeExpectancySubquery;
```

This code block calculates the average life expectancy of male astronauts. This could be useful for understanding the gender differences in life expectancy among astronauts.

```python
%%sql
SELECT Graduate_Major, COUNT(*) AS Number
FROM astronauts
GROUP BY Graduate_Major
ORDER BY Number DESC
LIMIT 10;
```

This code block groups the astronauts by their graduate major and counts the number of astronauts in each group, ordered by the number of astronauts in each group in descending order. This could be useful for identifying the most common graduate majors among astronauts.

```python
%%sql
SELECT
    COUNT(*) AS Number_of_Astronauts,
    SUM(CASE WHEN Undergraduate_Major IS NOT NULL THEN 1 ELSE 0 END) AS Astronauts_with_Undergraduate_Degrees,
    SUM(CASE WHEN Graduate_Major IS NOT NULL THEN 1 ELSE 0 END) AS Astronauts_with_Graduate_Degrees
FROM astronauts;
```

This code block counts the total number of astronauts, the number of astronauts with undergraduate degrees, and the number of astronauts with graduate degrees. This could be useful for understanding the educational background of astronauts.

```python
%%sql
SELECT
  CASE
    WHEN INSTR(Birth_Place, ',') > 0 THEN SUBSTR(Birth_Place, INSTR(Birth_Place, ',') + 1)
    ELSE Birth_Place
  END AS STATE,
  COUNT(*) AS Number
FROM astronauts
GROUP BY STATE
ORDER BY Number DESC
LIMIT 5;
```

This code block groups the astronauts by the state in which they were born and counts the number of astronauts in each group, ordered by the number of astronauts in each group in descending order. This could be useful for identifying the states that have produced the most astronauts.

```python
%%sql
SELECT
    ROUND(AVG(Space_Flights), 2) AS Average_Number_Of_Space_Flight,
    ROUND(AVG(Space_Walks), 2) AS Average_Number_Of_Space_Walks
FROM astronauts;
```

This code block calculates the average number of space flights and space walks for all astronauts. This could be useful for understanding the experience level of astronauts.
